```python
import pandas as pd
import matplotlib.pyplot as plt

# Load abusive and non-abusive data
abusive_data = pd.read_csv('/kaggle/input/ddsentiment/train_abusive.csv')
non_abusive_data = pd.read_csv('/kaggle/input/ddsentiment/train_nonabusive.csv')

# Assuming you have a 'label' column with values 0 or 1, and a 'timestamp' column
# Convert 'timestamp' to datetime format if it's not already
abusive_data['timestamp'] = pd.to_datetime(abusive_data['date'])
non_abusive_data['timestamp'] = pd.to_datetime(non_abusive_data['date'])

# Create additional columns for week, month, and year
abusive_data['week'] = abusive_data['timestamp'].dt.isocalendar().week
non_abusive_data['week'] = non_abusive_data['timestamp'].dt.isocalendar().week
abusive_data['month'] = abusive_data['timestamp'].dt.month
non_abusive_data['month'] = non_abusive_data['timestamp'].dt.month
abusive_data['year'] = abusive_data['timestamp'].dt.year
non_abusive_data['year'] = non_abusive_data['timestamp'].dt.year

# Group by week, month, and year and count abusive and non-abusive comments
abusive_weekly_counts = abusive_data.groupby(['week']).size()
non_abusive_weekly_counts = non_abusive_data.groupby(['week']).size()

abusive_monthly_counts = abusive_data.groupby(['month']).size()
non_abusive_monthly_counts = non_abusive_data.groupby(['month']).size()

abusive_yearly_counts = abusive_data.groupby(['year']).size()
non_abusive_yearly_counts = non_abusive_data.groupby(['year']).size()

# Plotting
# fig, axes = plt.subplots(3, 1, figsize=(12, 15))

# # Weekly Counts for Abusive and Non-Abusive
# abusive_weekly_counts.plot(kind='line', color='red', label='Abusive', ax=axes[0])
# non_abusive_weekly_counts.plot(kind='line', color='blue', label='Non-Abusive', ax=axes[0])
# axes[0].set_title('Weekly Counts',fontweight='bold')
# axes[0].set_xlabel('Week',fontweight='bold')
# axes[0].set_ylabel('Count',fontweight='bold')
# axes[0].legend()

# # Monthly Counts for Abusive and Non-Abusive
# abusive_monthly_counts.plot(kind='line', color='red', label='Abusive', ax=axes[1])
# non_abusive_monthly_counts.plot(kind='line', color='blue', label='Non-Abusive', ax=axes[1])
# axes[1].set_title('Monthly Counts',fontweight='bold')
# axes[1].set_xlabel('Month',fontweight='bold')
# axes[1].set_ylabel('Count',fontweight='bold')
# axes[1].legend()

# # Yearly Counts for Abusive and Non-Abusive
# abusive_yearly_counts.plot(kind='line', color='red', label='Abusive', ax=axes[2])
# non_abusive_yearly_counts.plot(kind='line', color='blue', label='Non-Abusive', ax=axes[2])
# axes[2].set_title('Yearly Counts',fontweight='bold')
# axes[2].set_xlabel('Year',fontweight='bold')
# axes[2].set_ylabel('Count',fontweight='bold')
# axes[2].legend()

# plt.tight_layout()
# plt.show()
import matplotlib.pyplot as plt
plt.rcParams['font.size'] = 10
plt.rcParams['font.weight'] = 'bold'

# Set font size
plt.rc('font', size=10)

# Create subplots with smaller figure size
fig, axes = plt.subplots(3, 1, figsize=(6, 10))
plt.grid(True)

# Weekly Counts for Abusive and Non-Abusive
abusive_weekly_counts.plot(kind='line', color='red', label='Abusive', ax=axes[0])
non_abusive_weekly_counts.plot(kind='line', color='blue', label='Non-Abusive', ax=axes[0])
axes[0].set_title('Weekly Counts', fontweight='bold')  # Bold title
axes[0].set_xlabel('Week', fontweight='bold')  # Bold x-axis label
axes[0].set_ylabel('Count', fontweight='bold')  # Bold y-axis label
axes[0].legend()
axes[0].grid(True)
axes[0].set_xticks(range(0, 60,5))  
#axes[0].set_aspect(0.5)
#axes[0].set_xticks(range(min(abusive_weekly_counts.index), max(abusive_weekly_counts.index) + 1, 1))

# axes[0].set_aspect(0.5)

# Monthly Counts for Abusive and Non-Abusive
abusive_monthly_counts.plot(kind='line', color='red', label='Abusive', ax=axes[1])
non_abusive_monthly_counts.plot(kind='line', color='blue', label='Non-Abusive', ax=axes[1])
axes[1].set_title('Monthly Counts', fontweight='bold')  # Bold title
axes[1].set_xlabel('Month', fontweight='bold')  # Bold x-axis label
axes[1].set_ylabel('Count', fontweight='bold')  # Bold y-axis label
axes[1].legend()
axes[1].grid(True)
axes[1].set_xticks(range(0, 12,1))
# axes[1].set_aspect(0.5)

# Yearly Counts for Abusive and Non-Abusive
abusive_yearly_counts.plot(kind='line', color='red', label='Abusive', ax=axes[2])
non_abusive_yearly_counts.plot(kind='line', color='blue', label='Non-Abusive', ax=axes[2])
axes[2].set_title('Yearly Counts', fontweight='bold')  # Bold title
axes[2].set_xlabel('Year', fontweight='bold')  # Bold x-axis label
axes[2].set_ylabel('Count', fontweight='bold')  # Bold y-axis label
axes[2].legend()
axes[2].grid(True)
axes[2].set_xticks(range(2013, 2019,1))
axes[2].set_aspect(0.01)





# Adjust tick parameters for readability
for ax in axes:
    ax.tick_params(axis='x', labelsize=8)  # Adjust x-axis tick label size
    ax.tick_params(axis='y', labelsize=8)  # Adjust y-axis tick label size
    
    
# plt.yticks(range(20, 85, 5), fontsize='large', fontweight='bold') 
    
# plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'abusive_data' and 'non_abusive_data' are your dataframes

# Convert 'time' column to datetime format
abusive_data['timestamp1'] = pd.to_datetime(abusive_data['time'], format='%H:%M:%S')
non_abusive_data['timestamp1'] = pd.to_datetime(non_abusive_data['time'], format='%H:%M:%S')

# Group by hour and count the occurrences
abusive_hourly_counts = abusive_data.groupby(abusive_data['timestamp1'].dt.hour).size()
non_abusive_hourly_counts = non_abusive_data.groupby(non_abusive_data['timestamp1'].dt.hour).size()

# Plotting

plt.figure(figsize=(8, 5))  # Adjusted figure size

# Plot abusive comments count
plt.plot(abusive_hourly_counts.index, abusive_hourly_counts.values, label='Abusive Comments', color='red')

# Plot non-abusive comments count
plt.plot(non_abusive_hourly_counts.index, non_abusive_hourly_counts.values, label='Non-Abusive Comments', color='blue')

plt.title('Distribution of Abusive and Non-Abusive Comments Over Each Hour', fontweight='bold')  # Bold title
plt.xlabel('Hour of the Day', fontweight='bold')  # Bold x-axis label
plt.ylabel('Number of Comments', fontweight='bold')  # Bold y-axis label
plt.legend(fontsize='large', loc='upper right').set_visible(True)  # Bold legend
plt.grid(True)
plt.xticks(range(24), fontsize='large', fontweight='bold')  # Set ticks for every hour and make them bold
plt.yticks(fontsize='large', fontweight='bold')  # Make y-axis ticks bold
 



max_count = max(max(abusive_hourly_counts), max(non_abusive_hourly_counts))
min_count = min(min(abusive_hourly_counts), min(non_abusive_hourly_counts))

# Set aspect ratio manually to achieve square grid lines
plt.gca().set_aspect(12 / (max_count - min_count))

# Set y-axis limits dynamically
plt.ylim(20, 85)

# Adjust y-ticks to have intervals of 10
plt.yticks(range(20, 85, 5), fontsize='large', fontweight='bold')  

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
abusive_religion_counts = (abusive_data['religion'] == 1).sum()
abusive_political_counts = (abusive_data['political'] == 1).sum()
abusive_other_counts = (abusive_data['other'] == 1).sum()


non_abusive_religion_counts = (non_abusive_data['religion'] == 1).sum()
non_abusive_political_counts = (non_abusive_data['political'] == 1).sum()
non_abusive_other_counts = (non_abusive_data['other'] == 1).sum()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['religion', 'political', 'other']
abusive_counts = [abusive_religion_counts, abusive_political_counts, abusive_other_counts]
non_abusive_counts = [non_abusive_religion_counts, non_abusive_political_counts, non_abusive_other_counts]

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))  # Adjusted figure size

bar_width = 0.35
index = range(len(categories))

plt.bar(index, abusive_counts, bar_width, label='Abusive', color='red')
plt.bar([i + bar_width for i in index], non_abusive_counts, bar_width, label='Non-Abusive')

plt.xlabel('Categories', fontweight='bold')  # Bold x-axis label
plt.ylabel('Counts', fontweight='bold')  # Bold y-axis label
plt.title('Counts of Categories for Abusive and Non-Abusive Data', fontweight='bold')  # Bold title
plt.xticks([i + bar_width/2 for i in index], categories, fontsize='large', fontweight='bold')  # Set x-axis ticks and make them bold
plt.yticks(fontsize='large', fontweight='bold')  # Make y-axis ticks bold
plt.legend(fontsize='large').set_visible(True)  # Bold legend

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()

```
